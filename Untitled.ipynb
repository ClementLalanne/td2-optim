# Generation of synthetic data

We start by importing the signal.

```python
true_x = np.loadtxt(PATH_TO_DATA+"x.txt")
```

Then we build the sampling grid.

```python
N = true_x.shape[0]
Tmin = 1
Tmax = 1000
T = Tmin * np.exp(- np.arange(N) * np.log(Tmin/Tmax) / (N-1))
```

The original signal is then:

```python
plt.plot(T, true_x)
plt.xscale('log')
plt.xlabel('T')
plt.ylabel('x')
plt.show()
```

Now we create the sampling strategy in the Laplace space.

```python
M = 50
tmin = 0
tmax = 1.5
t = tmin + np.arange(M)*(tmax-tmin)/(M-1)
```

And we build the transformation kernel.

```python
K = np.exp(- np.tensordot(t, T, axes=0))
```

We create some noisy data.

```python
sigma = 0.01
y = K.dot(true_x) + np.random.multivariate_normal(np.zeros(M), np.identity(M))
```

And we plot them.

```python
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

# Smoothness prior

The function 
$$f(x) = \frac{1}{2} ||Kx-y||^2 + \beta \frac{1}{2} ||Dx||^2$$
is differentiable with gradient 
$$\nabla f(x) = K^T(Kx-y) + \beta D^TDx$$
and hence this gradient is lipschitz continuous with for instance the lipschitz constant $||K^TK||_{fro} + \beta ||D^TD||_{fro}$.
So we can try a gradient descent with explicit learning rate.

```python
beta = 100
```

```python
D = np.identity(N)
for i in range(N):
    D[i, i-1] = - 1
```

```python
lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro') + beta * np.linalg.norm(D.T.dot(D), ord='fro')
lr = 1/lipschitz_grad_const
```

```python
def build_grad(x, y):
    return K.T.dot(K.dot(x)-y) + beta * D.T.dot(D.dot(x))
```

```python
def grad_desc(n_iter, lr = lr):
    ret = np.zeros((n_iter, N))
    ret[0] = np.random.randn(N)
    for i in range(1, n_iter):
        ret[i] = ret[i-1] - lr * build_grad(ret[i-1], y)
    return ret 
```

```python
seq = grad_desc(20000, lr=lr)
norm_err = np.linalg.norm(seq-true_x[np.newaxis, :], axis=1)**2/np.linalg.norm(true_x)**2
plt.xscale('log')
plt.yscale('log')
plt.plot(norm_err)
plt.show()
```

```python
plt.plot(T, seq[-1])
plt.xscale('log')
plt.xlabel('T')
plt.ylabel('recovered x')
plt.show()
```

However, we can also find a minimum by solving a linear system in order to find a point with zero grad.

```python
x_rec = np.linalg.solve(K.T.dot(K)+beta * D.T.dot(D), K.T.dot(y))
```

```python
plt.plot(T, x_rec)
plt.xscale('log')
plt.xlabel('T')
plt.ylabel('recovered x')
plt.show()
```

# Smoothness prior + constraints

The constraint added is a projection onto a convex set that is easily proximable. Hence we can use a projected gradient algorithm in order to find a sequence that converges weakly to the solution set.

```python
beta = 500
xmin = np.min(true_x)
xmax = np.max(true_x)
```

```python
lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro') + beta * np.linalg.norm(D.T.dot(D), ord='fro')
lr = 1/lipschitz_grad_const
lambd = 1 - lr * lipschitz_grad_const / 4
```

```python
def build_grad(x, y):
    return K.T.dot(K.dot(x)-y) + beta * D.T.dot(D.dot(x))
```

```python
def proj_grad_desc(n_iter, lr = lr, lambd = lambd):
    ret = np.zeros((n_iter, N))
    ret[0] = np.random.randn(N)
    for i in range(1, n_iter):
        projdual = np.minimum(np.maximum(ret[i-1] - lr * build_grad(ret[i-1], y), xmin), xmax)
        ret[i] = ret[i-1] + lambd * (projdual - ret[i-1])
    return ret 
```

```python
seq = proj_grad_desc(20000)
norm_err = np.linalg.norm(seq-true_x[np.newaxis, :], axis=1)**2/np.linalg.norm(true_x)**2
plt.xscale('log')
plt.yscale('log')
plt.plot(norm_err)
plt.show()
```

```python
plt.plot(T, seq[-1])
plt.xscale('log')
plt.xlabel('T')
plt.ylabel('recovered x')
plt.show()
```

# Sparsity prior

The constraint added is easily proximable. Hence we can use a forward-backward algorithm in order to find a sequence that converges weakly to the solution set.

```python
beta = 0.0001
```

```python
lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro') + beta * np.linalg.norm(D.T.dot(D), ord='fro')
lr = 1/lipschitz_grad_const
lambd = 1 - lr * lipschitz_grad_const / 4
```

```python
def build_grad(x, y):
    return K.T.dot(K.dot(x)-y) 
```

```python
def prox(x):
    return np.sign(x)* np.maximum(np.abs(x)-beta, 0)
```

```python
def forward_backward(n_iter, lr = lr, lambd = lambd):
    ret = np.zeros((n_iter, N))
    ret[0] = np.random.randn(N)
    for i in range(1, n_iter):
        proxdual = prox(ret[i-1] - lr * build_grad(ret[i-1], y))
        ret[i] = ret[i-1] + lambd * (proxdual - ret[i-1])
    return ret 
```

```python
seq = forward_backward(20000)
norm_err = np.linalg.norm(seq-true_x[np.newaxis, :], axis=1)**2/np.linalg.norm(true_x)**2
plt.xscale('log')
plt.yscale('log')
plt.plot(norm_err)
plt.show()
```

```python
plt.plot(T, seq[-1])
plt.xscale('log')
plt.xlabel('T')
plt.ylabel('recovered x')
plt.show()
```
