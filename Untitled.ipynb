```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw
```

```python
PATH_TO_DATA = ""
```

# Generation of synthetic data

We start by importing the signal.

```python
true_x = np.loadtxt(PATH_TO_DATA+"x.txt")
```

Then we build the sampling grid.

```python
N = true_x.shape[0]
Tmin = 1
Tmax = 1000
T = Tmin * np.exp(- np.arange(N) * np.log(Tmin/Tmax) / (N-1))
```

The original signal is then:

```python
plt.plot(T, true_x)
plt.xscale('log')
plt.xlabel('T')
plt.ylabel('x')
plt.show()
```

Now we create the sampling strategy in the Laplace space.

```python
M = 50
tmin = 0
tmax = 1.5
t = tmin + np.arange(M)*(tmax-tmin)/(M-1)
```

And we build the transformation kernel.

```python
K = np.exp(- np.tensordot(t, T, axes=0))
```

We create some noisy data.

```python
sigma = 0.01 * K.dot(true_x)[0]
y = K.dot(true_x) + sigma * np.random.multivariate_normal(np.zeros(M), np.identity(M))
```

And we plot them.

```python
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
def err(x_hat):
    return np.linalg.norm(x_hat - true_x, axis=-1)**2/np.linalg.norm(true_x)**2

def best_beta(algo, betas, **kwargs):
    es = []
    for beta in betas:
        es.append(err(algo(beta=beta, **kwargs)[0][-1]))
    return betas[np.array(es).argmin()]

def optimize_beta(algo, n_beta=100, bmin=-5, bmax=2, **kwargs):
    beta = best_beta(algo, 10**np.linspace(bmin, bmax, n_beta), **kwargs)
    return best_beta(algo, 10**np.linspace(np.log10(beta/2), np.log10(beta*2), n_beta), **kwargs)

reconstructions = {}
def evaluate(algo, beta=None, **kwargs):
    if beta is None:
        beta = optimize_beta(algo, **kwargs)
        print('Best beta: {:.2e}'.format(beta))
    seq, obj = algo(beta=beta, **kwargs)
    norm_err = err(seq)
    reconstructions[algo.__name__] = (beta, seq[-1])
    
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    if seq.shape[0] > 1:
        plt.xscale('log')
        plt.yscale('log')
        plt.plot(norm_err, label='Reconstruction error')
        plt.xlabel('Number of iterations')
        plt.ylabel('', color='C0')
        plt.tick_params(axis='y', colors='C0')
        plt.legend(loc='upper left')
        plt.twinx()
        plt.plot(obj, color='C1', label='Objective')
        plt.yscale('log')
        plt.ylabel('', color='C1')
        plt.tick_params(axis='y', colors='C1')
        plt.legend(loc='upper right')
    else:
        plt.bar(0, norm_err[0])
        plt.xlim(-2, 2)
        plt.ylabel('Reconstruction error')
    
    plt.subplot(1, 2, 2)
    plt.plot(T, true_x, label='true $x$')
    plt.plot(T, seq[-1], label='recovered $x$')
    plt.xscale('log')
    plt.xlabel('$T$')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

# Smoothness prior

The function 
$$f(x) = \frac{1}{2} ||Kx-y||^2 + \beta \frac{1}{2} ||Dx||^2$$
is differentiable with gradient 
$$\nabla f(x) = K^T(Kx-y) + \beta D^TDx$$
and hence this gradient is lipschitz continuous with for instance the lipschitz constant $||K^TK||_{fro} + \beta ||D^TD||_{fro}$.
So we can try a gradient descent with explicit learning rate.

```python
D = np.identity(N)
for i in range(N):
    D[i, i-1] = - 1
    
def build_grad(x, beta):
    return K.T.dot(K.dot(x)-y) + beta * D.T.dot(D.dot(x))
    
def grad_desc(beta, n_iter, **kwargs):
    lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro') + beta * np.linalg.norm(D.T.dot(D), ord='fro')
    lr = 1/lipschitz_grad_const
    ret = np.zeros((n_iter, N))
    ret[0] = np.zeros(N)
    for i in range(1, n_iter):
        ret[i] = ret[i-1] - lr * build_grad(ret[i-1], beta)
    obj = np.linalg.norm(ret.dot(K.T) - y, axis=1)**2/2 + beta * np.linalg.norm(ret.dot(D.T), axis=1)**2/2
    return ret, obj

evaluate(grad_desc, n_iter=20000, n_beta=20)
```

However, we can also find a minimum by solving a linear system in order to find a point with zero grad.

```python
def smoothness_prior(beta, **kwargs):
    return np.linalg.solve(K.T.dot(K) + beta * D.T.dot(D), K.T.dot(y))[None], None

evaluate(smoothness_prior)
```

# Smoothness prior + constraints

The constraint added is a projection onto a convex set that is easily proximable. Hence we can use a projected gradient algorithm in order to find a sequence that converges weakly to the solution set.

```python
xmin = np.min(true_x)
xmax = np.max(true_x)

def build_grad(x, beta):
    return K.T.dot(K.dot(x)-y) + beta * D.T.dot(D.dot(x))

def proj_grad_desc(beta, n_iter, **kwargs):
    lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro') + beta * np.linalg.norm(D.T.dot(D), ord='fro')
    lr = 1/lipschitz_grad_const
    lambd = 1 - lr * lipschitz_grad_const / 4
    ret = np.zeros((n_iter, N))
    ret[0] = np.zeros(N)
    for i in range(1, n_iter):
        projdual = np.clip(ret[i-1] - lr * build_grad(ret[i-1], beta), xmin, xmax)
        ret[i] = ret[i-1] + lambd * (projdual - ret[i-1])
    obj = np.linalg.norm(ret.dot(K.T) - y, axis=1)**2/2 + beta * np.linalg.norm(ret.dot(D.T), axis=1)**2/2
    return ret, obj

evaluate(proj_grad_desc, n_iter=20000, n_beta=20)
```

# Sparsity prior

The constraint added is easily proximable. Hence we can use a forward-backward algorithm in order to find a sequence that converges weakly to the solution set.

```python
def build_grad(x, y):
    return K.T.dot(K.dot(x)-y) 

def prox(x, gamma):
    return np.sign(x) * np.maximum(np.abs(x)- gamma, 0)

def forward_backward(beta, n_iter, **kwargs):
    lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro')
    lr = 1/lipschitz_grad_const
    lambd = 1 - lr * lipschitz_grad_const / 4
    ret = np.zeros((n_iter, N))
    ret[0] = np.zeros(N)
    for i in range(1, n_iter):
        proxdual = prox(ret[i-1] - lr * build_grad(ret[i-1], y), lr * beta)
        ret[i] = ret[i-1] + lambd * (proxdual - ret[i-1])
    obj = np.linalg.norm(ret.dot(K.T) - y, axis=1)**2/2 + beta * np.linalg.norm(ret, ord=1, axis=1)
    return ret, obj

evaluate(forward_backward, n_iter=20000, n_beta=20)
```

# Maximum entropy regularization

1. $\varphi$ is convex: its domain $\mathbb R_+$ is convex, $u \mapsto u \log u$ is convex on $\mathbb R_+^\ast$ as its derivative $u \mapsto 1 + \log u$ is non-decreasing, and $\varphi(u) \longrightarrow_{u\rightarrow0^+} 0$. Therefore, $\mathrm{ent}$ is convex as a sum of convex functions. It is proper, as $\mathrm{ent}(0) = 0 < +\infty$, and lower-semicontinuous because $\varphi$ is continuous on $\mathbb R_+$ which is closed and $+\infty$ elsewhere, and a sum of lower-semicontinuous functions is lower-semicontinuous. $\varphi$ is differentiable on $\mathbb R \setminus \{0\}$ but not in $0$, as it if were $1 + \log u \longrightarrow_{u \rightarrow 0^+} f'(u) = -\infty$ because of the derivative limit theorem. Therefore, as $\mathrm{ent}$ is separable, and differentiability implies the existence of partial derivatives, it is differentiable on $\{x \in \mathbb R^N \,|\, \lVert x \rVert_0 = N \}$.

2. The objective is strictly convex, as it is the sum of a convex and strictly convex functions ($\mathrm{ent}$ is strictly convex as $\varphi''(u) = \frac1u > 0$. Besides, the objective is coercive as $ \mathrm{ent}(x) \longrightarrow_{\lVert x \rVert \rightarrow +\infty} +\infty$. Therefore, the solution exists and is unique.

3. $\begin{align*}
\mathrm{prox}_{\gamma\varphi}(u) &= \inf_{v \geq 0} \underbrace{\frac12(u-v)^2 + \gamma v\log(v)}_{\triangleq f(v)}
\end{align*}$

$\begin{align*}
f'(v) &= v - u + \gamma\log v + \gamma \\
f'(v) = 0 &\iff v + \gamma\log v = u - \gamma \\
&\iff v e^{\frac v\gamma} = e^{\frac{u}\gamma-1} \\
&\iff \frac v\gamma = W(\frac1\gamma e^{\frac u\gamma-1}) \\
&\iff v = \gamma W(\frac1\gamma e^{\frac u\gamma-1})
\end{align*}$
The minimizer can't be $0$, as $f'(v) \longrightarrow_{v\rightarrow 0^+} -\infty$, so there exists an $\varepsilon > 0$ such that $f(\varepsilon) < f(0)$. Therefore it is in 
$]0,+\infty[$, and thus $\mathrm{prox}_{\gamma\varphi}(u) = \gamma W(\frac1\gamma e^{\frac u\gamma-1})$.

Finally, as $\mathrm{ent}$ is separable, $\mathrm{prox}_{\gamma\mathrm{ent}}(x) = \left( \mathrm{prox}_{\gamma\varphi}(x^{(n)})\right)$

4. As $\mathrm{ent}$ is proximable and $x\mapsto \frac12\lVert Kx-y\rVert^2$ is both proximable and differentiable, we can either use a forward-backward or a Douglas Rachford algorithm.

```python
def build_grad(x, y):
    return K.T.dot(K.dot(x)-y) 

def prox(x, gamma):
    return gamma * np.real(lambertw(np.exp(x/gamma - 1) / gamma))

def ent(x):
    e = np.zeros_like(x)
    np.log(x, where=x > 0, out=e)
    return np.sum(x * e + 1/np.e, axis=-1) # add 1/e so that it remains >= 0

def forward_backward_ent(beta, n_iter, **kwargs):
    lipschitz_grad_const = np.linalg.norm(K.T.dot(K), ord='fro')
    lr = 1/lipschitz_grad_const
    lambd = 1 - lr * lipschitz_grad_const / 4
    ret = np.zeros((n_iter, N))
    ret[0] = np.zeros(N)
    for i in range(1, n_iter):
        proxdual = prox(ret[i-1] - lr * build_grad(ret[i-1], y), lr * beta)
        ret[i] = ret[i-1] + lambd * (proxdual - ret[i-1])
    obj = np.linalg.norm(ret.dot(K.T) - y, axis=1)**2/2 + beta * ent(ret)
    return ret, obj

evaluate(forward_backward_ent, n_iter=200000, beta=1e-2)
```

```python
def prox2(x, gamma, inv):
    return inv.dot(x + gamma * K.T.dot(y))

def douglas_rachford(beta, n_iter, gamma=1e1, lambd=1, **kwargs):
    inv = np.linalg.inv(np.eye(N) + gamma * K.T.dot(K))
    ret = np.zeros((n_iter, N))
    ret[0] = np.zeros(N)
    x = np.zeros(N)
    for i in range(1, n_iter):
        p1 = prox(ret[i-1], gamma * beta)
        p2 = prox2(2*p1 - ret[i-1], gamma, inv)
        x += lambd * (p2 - p1)
        ret[i] = p2
    obj = np.linalg.norm(ret.dot(K.T) - y, axis=1)**2/2 + beta * ent(ret)
    return ret, obj

evaluate(douglas_rachford, n_iter=20000, beta=1e-2)
```

5. The Douglas-Rachford seems to converge much faster than the Forward-Backward algorithm, and produces better-looking results. However, they both converge to the same value objective (and reconstruction error).

```python
evaluate(forward_backward_ent, n_iter=200000, bmin=-1, n_beta=20)
evaluate(douglas_rachford, n_iter=20000, bmin=-1, n_beta=20)
```

7. In terms of reconstruction error, the best results come from the smoothness prior + constraints or the sparsity regularization. As the former also introduces sparsity through its constraint $x \geq 0$, it seems that sparsity is the key to good results.

8. The problem can be reformulated as: $$\min_{x\in\mathbb R^N} \mathrm{ent}(x) + \iota_{\mathcal B\left(y, \sqrt{\eta M\sigma^2}\right)}(Kx)$$
wich can be solved with the PPXA+ algorithm, since both operators are easily proximable and the objective is of the form $\sum_i g_i \circ L_i$.

```python
def prox3(z, r):
    return y + min(1, r / np.linalg.norm(z - y)) * (z - y)

def ppxa(beta, n_iter, lambd=1, gamma=10, **kwargs):
    eta = beta # to the outside, it is named beta in order to optimize for it
    r = np.sqrt(eta * M) * sigma
    inv = np.linalg.inv(np.eye(N) + K.T.dot(K))
    ret = np.zeros((n_iter, N))
    ret[0] = np.zeros(N)
    x1, x2 = np.zeros(N), np.zeros(M)
    for i in range(1, n_iter):
        y1, y2 = prox(x1, gamma), prox3(x2, r)
        c = inv.dot(y1 + K.T.dot(y2))
        x1, x2 = x1 + lambd * (2*c - ret[i-1] - y1), x2 + lambd * (K.dot(2*c - ret[i-1]) - y2)
        ret[i] = ret[i-1] + lambd * (c - ret[i-1])
    obj = ent(ret)
    return ret, obj

evaluate(ppxa, n_iter=200000, beta=1e-1)
```
